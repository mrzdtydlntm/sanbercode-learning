```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import f1_score, make_scorer, roc_auc_score
from sklearn.pipeline import Pipeline
from sklearn import decomposition, datasets
```

```python
x = df_train_new.drop('price_range_encode', axis = 1)
y = df_train_new['price_range_encode']
parameter = {'n_neighbors' : np.arange(5,50), 'weights' : ['uniform', 'distance']}
```

```python
scoring = make_scorer(f1_score, average = 'micro')
```

```python
gscv = GridSearchCV(KNeighborsClassifier(), parameter, cv = 5, scoring = scoring)
```

```python
gscv.fit(x,y)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
x_tes = df_test_new
y_pred = gscv.predict(x_tes)
```

```python
y_pred
```

```python
subm = pd.DataFrame({
    'index':df_test['index'],
    'price_range':y_pred
})

subm.to_csv('price_range_pred.csv', index = False)
```

```python
pipe = Pipeline(steps=[('std_slc', StandardScaler()),
                           ('pca', decomposition.PCA()),
                           ('dec_tree', DecisionTreeClassifier())])
```

```python
n_components = list(range(1,x.shape[1]+1,1))
```

```python
criterion = ['gini', 'entropy']
max_depth = np.arange(2,20)
```

```python
parameters = dict(pca__n_components=n_components,
                      dec_tree__criterion=criterion,
                      dec_tree__max_depth=max_depth)
```

```python
clf_GS = GridSearchCV(pipe, parameters)
```

```python
clf_GS.fit(x, y)
```

```python
clf_GS.best_params_
```

```python
clf_GS.best_score_
```
