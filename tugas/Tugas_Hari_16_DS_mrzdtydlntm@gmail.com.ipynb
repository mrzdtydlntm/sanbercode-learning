* Machine Learning = pembelajaran dari mesin komputer yang improved dari data yang diproses secara mandiri

* feature = data yang dapat diukur dari suatu kejadian yang dapat diamati

target = data yang ingin di pelajari lebih dalam menggunakan ML

* Supervised = data yang memiliki input dan output ter 'klasifikasi' yaitu perintah untuk mengerjakan sesuai input dan output

Unsupervised = data yang tidak memiliki input dan output yang diinginkan, jadi mesin akan mengklasifikasikan sesuai 'kemiripan' data yang didapatkan

* Jenis supervised learning:

Regresi = dari korelasi input dari interpretasi probabilistik data training

Classification = data di 'klasifikasi' yaitu dikelompokkan menjadi kelompok tertentu

Decision Tree = menggunakan kontrol bersyarat yang akan membentuk suatu decision dan probabilitas konsekuensinya menjadikan seperti pohon

Random forest = membuat tree decision yang banyak kemudian mencari probabilitas terbaiknya

Neural networks =  mengelompokkan raw data, mengenali pola, atau menafsirkan data inputan

SVM = membuat statistika model berdasar pengklasifikasian

* Hyperparameter = parameter yang memengaruhi data output namun tidak berubah saat training data, karena diberikan di awalpendefinisian

Parameter = data acak untuk melatih data input sefleksibel mungkin agar sesuai dengan domain output yang tepat

* langkah2 membuat ML: memilih algoritma yang akan di terapkan kedalam model berdasar problem, menentukan hyperparaemternya dan mempelajari data (train) dengan fitting data dan mengaplikasikan modelnya

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
xtrain = x.reshape(-1, 1)
ytrain = y.reshape(-1, 1)
xtes = x_new.reshape(-1, 1)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor(max_depth = 5, min_samples_split = 2, min_samples_leaf = 5, max_leaf_nodes = 5)
```

```python
dtree.fit(xtrain,ytrain)
```

```python
ypred = dtree.predict(xtes)
```

```python
ytrain = dtree.predict(xtrain)
```

```python
plt.scatter(x,y, label = 'Data Asli')
plt.scatter(xtrain, ytrain, label = 'Prediksi data training')
plt.plot(xtes, ypred, label = 'Prediksi data baru')
plt.legend()
plt.show()
```
