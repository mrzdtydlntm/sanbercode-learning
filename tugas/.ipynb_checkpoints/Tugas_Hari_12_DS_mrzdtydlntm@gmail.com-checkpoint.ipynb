Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
plt.plot(x, cos, color='red', label='Cosx')
plt.plot(x, cos3, color='green', linestyle='--', label='3 Cosx')
plt.plot(x, sin, color='blue', linestyle='-.', label='Sin(x-90)')
plt.plot(x, sin2, color='yellow', linestyle='', marker='x', label='2 Sinx')
plt.legend(loc='center', frameon=False)
plt.title('Grafik Cos Sin')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.show()
```
